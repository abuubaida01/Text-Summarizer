# **Function**

```python
## let's make a function
def TextSummarizer(link):
    #will get ID
    ID = link.split("=")[1]

    #will get caption 
    transcript = YouTubeTranscriptApi.get_transcript(ID)

    #let's gather text 
    text = " ".join([i['text'] for i in transcript])

    #let's summarize it by Transformer
    iter = int(len(text)/1000)  ## as words are in thousands, so each time I will summarize 1K

    summary_text= []
    for i in range(0, iter+1):
        start = i * 1000 # it will start from 0, so =0
        end = (i+1)* 1000 # it will end on 1000
        summary = summarizer(text[start:end])
        summary = summary[0]['summary_text']
        summary_text.append(summary)
        
        #as It will be list of dict, at position 0, we have dict, we will get key = summary_text
    return ''.join(summary_text)
    
```
